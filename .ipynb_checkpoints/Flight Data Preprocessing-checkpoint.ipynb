```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_b = pd.read_csv("business.csv")
df_e = pd.read_csv("economy.csv")
```

```python
print(df_b.shape,df_e.shape)
```

```python
df_b.head()
```

```python
df_b["class"] = 'business'
df_e["class"] = 'economy'
```

```python
df = pd.concat([df_e, df_b], axis=0)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
import re
df['stop'] = df['stop'].apply(lambda x: x.replace('non-stop', '0'))
df['stop'] = df['stop'].apply(lambda x: re.sub(r'\D', '', x))
```

```python
df['stop'].unique()
```

```python
df.head()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
col = df.select_dtypes(include=['object'])
for i in col:
    print(df[i].value_counts())
    print("------------------------------")
```

<p>early_morning: 4am-7am<br>
morning: 7am-12pm<br>
afternoon: 12pm-6pm<br>
evening: 6pm-9pm<br>
night: 9pm-3am</p>

```python
df['dep'] = df['dep_time'].apply(lambda x: x[0:2])
df['arr'] = df['arr_time'].apply(lambda x: x[0:2])
```

```python
df.head()
```

```python
df['dep'] = pd.to_numeric(df['dep'])
df['arr'] = pd.to_numeric(df['arr'])
```

```python
df.info()
```

```python
df['Departure'] = df['dep'].apply(lambda x:"Early_Morning" if x in range(4,7) else "Morning" if x in range(7,12) else "Afternoon" if x in range(12,18) else "Evening" if x in range(18,20) else "Night" )
df['arrival'] = df['arr'].apply(lambda x:"Early_Morning" if x in range(4,7) else "Morning" if x in range(7,12) else "Afternoon" if x in range(12,18) else "Evening" if x in range(18,20) else "Night" )
```

```python
df['Departure'].unique()
```

```python
df.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
df['time'] = df['time_taken'].apply(lambda x: x.replace('h','').replace('m','').replace(' ',':'))
```

```python
df['time'].head()
```

```python
df[['hr', 'mins']] = df['time'].str.split(':', expand=True)
df['hr'] = pd.to_numeric(df['hr'])
df['mins'] = pd.to_numeric(df['mins'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['mins'] = df['mins'].apply(lambda x: round(x / 60, 2) if pd.notna(x) else 0)
```

```python
df['time'] = df['hr']+df['mins']
```

```python
df.isnull().sum()
```

```python
df = df.drop(['airline','dep_time','arr_time','dep','arr','time_taken','hr','mins'],axis=1)
```

```python
df.head()
```

```python
df['price'] = df['price'].apply(lambda x: x.replace(',',''))
df['price'] = pd.to_numeric(df['price'])
```

```python
df['stop'] = pd.to_numeric(df['stop'])
```

```python
df.info()
```

```python
import datetime as dt
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%d-%m-%Y', errors = 'coerce')
```

```python
df.head()
```

```python
df['date1'] = df['date'].dt.day
df['month'] = df['date'].dt.month
```

```python
df = df.drop(['date'],axis=1)
df = df.rename(columns={'date1':'date'})
df['month'] = pd.to_numeric(df['month'])
df['date'] = pd.to_numeric(df['date'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df[[ 'date', 'month','ch_code', 'num_code','class','from','Departure','stop','to',
        'arrival','time','price']]
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```
